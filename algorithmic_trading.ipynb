```python
# Remove the volume
data.drop('Volume', axis=1)
```

```python
# Read data from CSV file into a Pandas dataframe
data = pd.read_csv('eurusd_1h.csv')

# Calculate the low and high of the last candle
last_candle_low = data['Low'].shift(1)
last_candle_high = data['High'].shift(1)

# Calculate the low and high of the current candle
current_candle_low = data['Low']
current_candle_high = data['High']

# Initialize the buy and sell signal columns to 0
data['Buy Signal'] = np.where(current_candle_high > last_candle_high, 1, 0)
data['Sell Signal'] = np.where(current_candle_low < last_candle_low, 1, 0)

# Print the updated Pandas dataframe
print(data)
```
